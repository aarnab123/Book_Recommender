```python
books = pd.read_csv('/kaggle/input/recommendation-system-data/BX-Books.csv',sep=';',on_bad_lines='skip',encoding='latin-1')
books
```

```python
books.columns
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
books.head()
```

```python
books.rename(columns={
    'Book-Title':'Title',
    'Book-Author':'Author',
    'Year-Of-Publication':'Year',
    'Image-URL-L':'Url'
},inplace=True)

books.head()
```

```python
user = pd.read_csv('/kaggle/input/recommendation-system-data/BX-Users.csv',sep=';',on_bad_lines='skip',encoding='latin-1')
user
```

```python
user.shape
```

```python
ratings = pd.read_csv('/kaggle/input/recommendation-system-data/BX-Book-Ratings.csv',sep=';',on_bad_lines='skip',encoding='latin-1')
ratings.head()
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'Book-Rating':'rating'
},inplace=True)
```

```python
print(books.shape)
print(user.shape)
print(ratings.shape)
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts()
```

**Extracting the users who rated more t han 100 books**

```python
X = ratings['user_id'].value_counts() >100
X[X].shape
```

**index positions of those users who rated more than 100 books**

```python
y = X[X].index
y
```

**Extracting the users from the ratings dataset**

```python
ratings= ratings[ratings['user_id'].isin(y)]
ratings
```

**Marging with books to identify the books corrosponding to the user with ratings on.**

```python
books.head(1)
```

```python
ratings_Books = ratings.merge(books, on='ISBN',)
ratings_Books.head()
```

```python
ratings_Books.shape
```

**Finding out how many times a book was rated by users**

```python
num_rating = ratings_Books.groupby('Title')['rating'].count().reset_index()
num_rating.head()
```

```python
num_rating.rename(columns={
    'rating':'no_rating'
},inplace=True)
```

```python
num_rating
```

**Creating the final dataset by marging with ratings_books data**

```python
final_data = ratings_Books.merge(num_rating,on='Title')
final_data
```

**Selecting those books that have been rated more than 40 times.**

```python
final_data = final_data[final_data['no_rating']>=40]
final_data
```

```python
final_data.drop_duplicates(['user_id','Title'],inplace=True)
final_data.shape
```

```python
book_pivot = final_data.pivot_table(columns='user_id',index='Title',values='rating')
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute', metric='euclidean')
model.fit(book_sparse)
```

```python
distance, suggestions = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
distance
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```
